```python
import matplotlib.pyplot as plt
```

# Scatter Plot

```python
day = [1, 2, 3, 4, 5]
axis_y = [23, 54, 68, 85, 64]

# adding colors
colors = ['r', 'g', 'b', 'm', 'y']
plt.scatter(day, axis_y, color=colors, s=60)

plt.xlabel("day")
plt.ylabel("number")
plt.title("This is a title")

plt.show()
```

```python
# adding separate sizes and tranparency->alpha

day = [1, 2, 3, 4, 5]
axis_y = [23, 54, 68, 85, 64]

# adding colors
colors = ['r', 'g', 'b', 'm', 'y']
size=[500, 100, 300, 800, 300]
plt.scatter(day, axis_y, color=colors, s=size, alpha=0.2)

plt.xlabel("day")
plt.ylabel("number")
plt.title("This is a title")

plt.show()
```

### Markers

```python
# adding separate sizes and tranparency->alpha

day = [1, 2, 3, 4, 5]
axis_y = [23, 54, 68, 85, 64]

# adding colors
colors = ['r', 'g', 'b', 'm', 'y']
size=[500, 100, 300, 800, 300]
plt.scatter(day, axis_y, color=colors, s=size, marker='*')

#list of some markers available
''' 
o	Circle
s	Square
^	Triangle Up
v	Triangle Down
D	Diamond
*	Star
'''

plt.xlabel("day")
plt.ylabel("number")
plt.title("This is a title")

plt.show()
```

```python

import numpy as np

x = np.random.rand(20) * 10
y = 2*x + np.random.randn(20) * 2  # Linear relationship with noise

# Create the scatter plot
plt.scatter(x, y, label='Data Points')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
