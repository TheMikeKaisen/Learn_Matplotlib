```python
import matplotlib.pyplot as plt
```

# Scatter Plot

```python
day = [1, 2, 3, 4, 5]
axis_y = [23, 54, 68, 85, 64]

# adding colors
colors = ['r', 'g', 'b', 'm', 'y']
plt.scatter(day, axis_y, color=colors, s=60)

plt.xlabel("day")
plt.ylabel("number")
plt.title("This is a title")

plt.show()
```

```python
# adding separate sizes and tranparency->alpha

day = [1, 2, 3, 4, 5]
axis_y = [23, 54, 68, 85, 64]

# adding colors
colors = ['r', 'g', 'b', 'm', 'y']
size=[500, 100, 300, 800, 300]
plt.scatter(day, axis_y, color=colors, s=size, alpha=0.2)

plt.xlabel("day")
plt.ylabel("number")
plt.title("This is a title")

plt.show()
```

### Markers

```python
# adding separate sizes and tranparency->alpha

day = [1, 2, 3, 4, 5]
axis_y = [23, 54, 68, 85, 64]

# adding colors
colors = ['r', 'g', 'b', 'm', 'y']
size=[500, 100, 300, 800, 300]
plt.scatter(day, axis_y, color=colors, s=size, marker='*')

#list of some markers available
''' 
o	Circle
s	Square
^	Triangle Up
v	Triangle Down
D	Diamond
*	Star
'''

plt.xlabel("day")
plt.ylabel("number")
plt.title("This is a title")

plt.show()
```

```python
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 1]

x2 = [1, 2, 3, 4, 5]
y2 = [4, 1, 6, 2, 8]

plt.scatter(x1, y1, color='blue', label='Dataset 1', marker='o')
plt.scatter(x2, y2, color='green', label='Dataset 2', marker='s')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Data Sets Scatter Plot')
plt.legend()
plt.show()
```
