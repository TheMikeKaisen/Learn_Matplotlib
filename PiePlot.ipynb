```python
import matplotlib.pyplot as plt
```

```python
x = [10, 20, 30, 23]
y = ["cpp", 'java', 'python', 'go']

plt.pie(x, labels=y)
plt.show()
```

```python
# explode
x = [10, 20, 30, 23]
y = ["cpp", 'java', 'python', 'go']
ex = [0.3, 0, 0, 0.1]

plt.pie(x, labels=y, explode=ex)
plt.show()
```

```python
# colors
x = [10, 20, 30, 23]
y = ["cpp", 'java', 'python', 'go']
ex = [0.3, 0, 0, 0.1]
colors = ['r', 'b', 'm', 'y']

plt.pie(x, labels=y, explode=ex, colors=colors)
plt.show()
```

```python
# adding percentage and shadow
x = [10, 20, 30, 23]
y = ["cpp", 'java', 'python', 'go']
ex = [0.3, 0, 0, 0.1]

plt.pie(x, labels=y, explode=ex, shadow=True, autopct= "%0.1f%%") # o.2f for 2 floating points
plt.show()
```

```python
# adding radius
x = [10, 20, 30, 23]
y = ["cpp", 'java', 'python', 'go']

plt.pie(x, labels=y, radius=0.5)
plt.show()
```

```python
# distance of labels from chart
x = [10, 20, 30, 23]
y = ["cpp", 'java', 'python', 'go']

plt.pie(x, labels=y, labeldistance=1.3) # labeldistance < 1 will set teh labels inside the chart
plt.show()
```

# startangle, textprops, wedgeprops

```python
x = [10, 20, 30, 23]
y = ["cpp", 'java', 'python', 'go']

# startangle -> used for rotating the pie chart
# textprops -> dictionary that contains text properties
# wedgeprops -> dictionary that contains edge properties
plt.pie(x, labels=y, labeldistance=1.3, startangle=90, textprops={"fontsize":15},
        wedgeprops={"edgecolor":"orange", "linewidth":4}) # labeldistance < 1 will set teh labels inside the chart
plt.show()
```

```python
x = [10, 20, 30, 23]
y = ["cpp", 'java', 'python', 'go']

plt.pie(x, labels=y, labeldistance=1.2, ) 

plt.title("This is an amazing title", fontsize=20)
plt.legend(loc=3)

plt.show()
```

# Creating a Donut PieChart

```python
x = [10, 20, 30, 23]
y = ["cpp", 'java', 'python', 'go']

plt.pie(x, labels=y, radius=1.5, shadow=True)
cr = plt.Circle(xy=(0,0), radius=0.8, facecolor='w') # radius of circle should always be lesser that radius of pie
plt.gca().add_artist(cr)


plt.show()
```
