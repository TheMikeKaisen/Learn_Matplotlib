# Bar Plot

```python
import matplotlib.pyplot as plt
```

```python
# creating a basic bar plot
axis_x = ['python', 'cpp', 'javascript', 'golang']
axis_y = [85, 65, 43, 88]

# initialize the bar
plt.bar(axis_x, axis_y)

# show the graph
plt.show()
```

```python
# creating horizontal bar graph
axis_x = ['python', 'cpp', 'javascript', 'golang']
axis_y = [85, 65, 43, 88]

# initialize the bar
plt.barh(axis_x, axis_y)

# show the graph
plt.show()
```

### Title, Labels, Fontsizes

```python
# adding title, x label, y label, and fontsizes to them

axis_x = ['python', 'cpp', 'javascript', 'golang']
axis_y = [85, 65, 43, 88]

plt.bar(axis_x, axis_y)
plt.xlabel("Language", fontsize=15)
plt.ylabel("no. of students")
plt.title("Language vs Number of student", fontsize=18)

plt.show()
```

### Color, Width

```python
axis_x = ['python', 'cpp', 'javascript', 'golang']
axis_y = [85, 65, 43, 88]

plt.xlabel("Language", fontsize=15)
plt.ylabel("no. of students")
plt.title("Language vs Number of student", fontsize=18)

#add color
barColor = ['b', 'y', 'm', 'r']
plt.bar(axis_x, axis_y, color=barColor, width = 0.1) #  0 <= width <= 1

plt.show()
```

### Edges - edgecolor, linewidth, linestyle

```python
axis_x = ['python', 'cpp', 'javascript', 'golang']
axis_y = [85, 65, 43, 88]

plt.xlabel("Language", fontsize=15)
plt.ylabel("no. of students")
plt.title("Language vs Number of student", fontsize=18)

barColor = ['b', 'y', 'm', 'r']

# add edge properties
plt.bar(axis_x, axis_y, color=barColor, width = 0.1, edgecolor='orange', linewidth=4, linestyle="-") 
# linestyle=> {":"="dotted", "-" = "solid" } 

plt.show()
```

### alpha - adjusts opacity of the graph

```python
axis_x = ['python', 'cpp', 'javascript', 'golang']
axis_y = [85, 65, 43, 88]

plt.xlabel("Language", fontsize=15)
plt.ylabel("no. of students")
plt.title("Language vs Number of student", fontsize=18)

barColor = ['b', 'y', 'm', 'r']


plt.bar(axis_x, axis_y, color=barColor, width = 0.1, edgecolor='orange', linewidth=4, alpha=0.5) # 0 < alpha < 1

plt.show()
```

### Legend -> label

```python
axis_x = ['python', 'cpp', 'javascript', 'golang']
axis_y = [85, 65, 43, 88]

plt.xlabel("Language", fontsize=15)
plt.ylabel("no. of students")
plt.title("Language vs Number of student", fontsize=18)

plt.bar(axis_x, axis_y, width=0.3, label="langauges")
plt.legend()

plt.show()
```

```python
# multiple graphs -> provides overlapped graphs
axis_x = ['python', 'cpp', 'javascript', 'golang']
axis_y1 = [85, 14, 43, 55]
axis_y2 = [34, 43, 18, 80]

plt.xlabel("Language", fontsize=15)
plt.ylabel("no. of students")
plt.title("Language vs Number of student", fontsize=18)

# graph will be overlapping here.
plt.bar(axis_x, axis_y1, color='red', width=0.3, label="langauge1")
plt.bar(axis_x, axis_y2, color="blue", width=0.3, label="langauge2")
plt.legend()

plt.show()
```
